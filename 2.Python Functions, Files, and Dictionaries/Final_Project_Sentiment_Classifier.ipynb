```python

punctuation_chars = ["'", '"', ",", ".", "!", ":", ";", '#', '@']

def strip_punctuation(wrd):
    for i in punctuation_chars:
        wrd = wrd.replace(i, "")
    return wrd

# lists of positive words to use
positive_words = []
with open("positive_words.txt") as pos_f:
    for lin in pos_f:
        if lin[0] != ';' and lin[0] != '\n':
            positive_words.append(lin.strip())
# function for counting positive words       
def get_pos(a):
    a = strip_punctuation(a)
    pscore = 0
    #for i in positive_words:
        #if i.lower() == positive_words:
            #pscore += 1
    for i in a.split():
        if i.lower() in positive_words:
            pscore += 1
    return pscore            

# lists of negative to use words to use          
negative_words = []
with open("negative_words.txt") as pos_f:
    for lin in pos_f:
        if lin[0] != ';' and lin[0] != '\n':
            negative_words.append(lin.strip())
# function for counting negative words 
def get_neg(a):
    s = 0
    a = strip_punctuation(a)
    for i in a.split():
        if i in negative_words:
            s += 1
    return s

fileconnection = open("project_twitter_data.csv", 'r')
lines = fileconnection.readlines()
fileconnection.close()

s = []

header = lines[0]
field_names = header.strip().split(',')

#for row in lines[1:]:
#    vals = row.strip().split(',')
#    #print(vals)
#    c = strip_punctuation(vals[0])
#    #print(c)
#    s.append(c)

print(len(s))    
#the text of a tweet
#the number of retweets of that tweet, 
#and the number of replies to that tweet        

outfile = open("resulting_data.csv", "w")
outfile.write('Number of Retweets, Number of Replies, Positive Score, Negative Score, Net Score')
outfile.write('\n')
# output each of the rows:
for row in lines[1:]:
    lst = row.strip().split(',')
    #outfile.write("{lst[1]}, {lst[2]}, {}, {}, {}".format(lst[1],lst[2],get_pos(lst[0]),get_neg(lst[0]),get_pos(lst[0])-get_neg(lst[0])))
    outfile.write("{}, {}, {}, {}, {}".format(lst[1], lst[2], get_pos(lst[0]), get_neg(lst[0]), (get_pos(lst[0])-get_neg(lst[0]))))
    #print(row)
    outfile.write('\n')
outfile.close()    

```
